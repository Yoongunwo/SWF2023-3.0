```python
total_df = pd.read_csv("book_data/공공도서관_도서_목록_2023_06.csv", encoding="utf-8")
```

```python
# 불필요한 컬럼 및 결측치, 중복된 데이터 drop
select_columns = ['AUTHR_NM', 'CL_SMBL_NO', 'PBLICTE_YEAR', 'TITLE_NM', 'ISBN_THIRTEEN_NO']

selected_df = total_df[select_columns].rename(columns={'AUTHR_NM' : '저작자', 'CL_SMBL_NO' : '분류기호번호', 'PBLICTE_YEAR': '발행년도', 'TITLE_NM' : '제목','ISBN_THIRTEEN_NO' :'ISBN'})
selected_df.dropna(inplace=True)
selected_df = selected_df.drop_duplicates(keep='first')
```

```python
selected_df
```

```python
# '분류기호번호', 'ISBN' 의 값이 numeric이 아닌 데이터 drop -> index 재설정
selected_df['분류기호번호'] = pd.to_numeric(selected_df['분류기호번호'], errors='coerce', downcast='integer')
selected_df['발행년도'] = pd.to_numeric(selected_df['발행년도'], errors='coerce', downcast='integer')
selected_df['ISBN'] = pd.to_numeric(selected_df['ISBN'], errors='coerce')

# Drop rows with NaN values in 'Classification Code Number' and 'ISBN' columns
selected_df.dropna(subset=['분류기호번호', '발행년도', 'ISBN'], inplace=True)

# Optional: Reset the DataFrame index
selected_df.reset_index(drop=True, inplace=True)
```

```python
# numeric으로 변경한 '분류기호번호', 'ISBN'을 str로 re-formatting
selected_df['분류기호번호'] = selected_df['분류기호번호'].apply(lambda x : str(x)[0:3])
selected_df['ISBN'] = selected_df['ISBN'].apply(lambda x : str(x)[0:13])

integers_array = np.arange(1000, 20001, 100)
selected_df['구매가격'] = np.random.choice(integers_array, size=len(selected_df), replace=True)
selected_df['2주일 대여 가격'] = selected_df['구매가격'] * 0.10
selected_df['1달 대여 가격'] = selected_df['구매가격'] * 0.20
selected_df['3달 대여 가격'] = selected_df['구매가격'] * 0.50
```

```python
selected_df
```

```python
KDC = {
    '000' : '총류',
    '100' : '철학',
    '200' : '종교',
    '300' : '사회과학',
    '400' : '자연과학',
    '500' : '기술과학',
    '600' : '예술',
    '700' : '언어',
    '800' : '문학',
    '900' : '역사',
}
```

```python
selected_df['장르'] = selected_df['분류기호번호'].apply(lambda x : KDC[x[0] + '00'] )
```

```python
column_order = ['저작자','제목', '발행년도', 'ISBN', '분류기호번호', '장르', '구매가격', '2주일 대여 가격', '1달 대여 가격', '3달 대여 가격']

# Rearrange the columns based on the desired order
selected_df = selected_df[column_order]
```

```python
selected_df.to_csv("book_data/total_mongo_db_book.csv", index=False)
```
