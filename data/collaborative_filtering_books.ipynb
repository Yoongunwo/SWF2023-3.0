```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
```

```python
# 방법2 - HOST, PORT
client = MongoClient(host='localhost', port=27017)
```

```python
print(client.list_database_names())
```

```python
# 방법2
db = client['REAL']
```

```python
# 방법2
collection = db['user']
```

```python
# # Retrieve all documents in the collection
# all_documents = collection.find()

# all_data = list(all_documents)
```

```python
all_data = list(collection.find({}, {'_id': 0}))  # Exclude '_id' field from the query
df = pd.DataFrame(all_data)
```

```python
df
```

```python
# '제목' 컬럼이 서로 다른 200개의 책을 뽑아서 20000개의 행을 구함
unique_titles = df['제목'].unique()
random_titles = pd.Series(unique_titles).sample(n=200, random_state=42).tolist()
choosen_books_df = df[df['제목'].isin(random_titles)].sample(n=20000, replace=True, random_state=42)
choosen_books_df
```

```python
choosen_books_df['나이대'] = [ np.random.randint(1,7)*10 for _ in range(20000) ]
choosen_books_df['성별'] = [ np.random.randint(0,2) for _ in range(20000) ] 
choosen_books_df['구독여부'] = [ np.random.randint(0,2) for _ in range(20000) ]
choosen_books_df['구매 및 대여기간'] = [np.random.choice(['2주일', '1달', '3달','구매']) for _ in range(20000)]
choosen_books_df['평점'] = [ np.random.randint(1,6) for _ in range(20000)]
```

```python
choosen_books_df['성별'] = choosen_books_df['성별'].apply(lambda x: '남자' if x ==1 else '여자')
```

```python
choosen_books_df['구독여부'] = choosen_books_df['구독여부'].apply(lambda x: 'Y' if x ==1 else 'N')
```

```python
choosen_books_df = choosen_books_df.reset_index(drop=True)
choosen_books_df
```

```python
grouped = choosen_books_df.groupby('제목')

for group_key, rows in grouped:
    rating_mean = np.mean(rows['평점'])
    choosen_books_df.loc[choosen_books_df['제목'] == group_key, '평균평점'] = round(rating_mean, 2)
```

```python
choosen_books_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 사용자가 입력한 컬럼들만 선택
user_age = 30
user_gender = '여자'
user_is_subscription = 'Y'

# 책들의 정보와 유사한 사용자들을 찾기 위해 나이대, 성별, 구독여부를 필터링하여 동일한 사용자들의 데이터를 추출
similar_users = choosen_books_df.loc[
    (choosen_books_df['나이대'] == user_age) &
    (choosen_books_df['성별'] == user_gender) &
    (choosen_books_df['구독여부'] == user_is_subscription)
]

# 콘텐츠 기반 필터링을 위해 도서 제목을 벡터화
vectorizer = CountVectorizer()
title_matrix = vectorizer.fit_transform(choosen_books_df['제목'])

# 코사인 유사도 계산
cosine_sim = cosine_similarity(title_matrix)

# 유사한 사용자들이 선호하는 책들 중에서 사용자가 평점을 매긴 책과 유사도가 가장 높은 책을 추천
recommended_books = []

for book in similar_users['제목']:
    book_idx = choosen_books_df[choosen_books_df['제목'] == book].index[0]
    sim_scores = list(enumerate(cosine_sim[book_idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  # 자기 자신을 제외하고 상위 5개 유사도 책들을 선택
    book_indices = [i[0] for i in sim_scores]
    recommended_books.extend(choosen_books_df['제목'].iloc[book_indices])

# 중복된 책들을 제거하고 인기도가 높은 순서대로 정렬하여 최종 추천 책을 선택
recommended_books = list(set(recommended_books))
recommended_books = choosen_books_df[choosen_books_df['제목'].isin(recommended_books)].sort_values(by='평균평점', ascending=False)[['제목', '평점']].reset_index(drop=True)
```

```python
unique_books_df = recommended_books.drop_duplicates(subset='제목', keep='first')
unique_books_df
```

```python
# 상위 10개의 추천 도서만 선택
top_10_recommended_books = unique_books_df.head(10)['제목'].values
```

```python
print("10개 추천 도서:")
for book in top_10_recommended_books:
    print(book)
```
